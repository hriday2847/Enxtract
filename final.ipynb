```python
import requests
import numpy as np
import cv2
from paddleocr import PaddleOCR

ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Function to load the image from a URL
def load_image_from_url(url):
    print(f"URL of image is: {url}")
    response = requests.get(url)
    image_data = np.asarray(bytearray(response.content), dtype="uint8")
    image = cv2.imdecode(image_data, cv2.IMREAD_COLOR)
    return image

# Function to preprocess the image, extract text using PaddleOCR
def preprocess_and_extract_text(image):
    # Perform OCR on the image
    result = ocr.ocr(image)

    words = []

    # Ensure result is not None, an empty list, or contains 'None' entries
    if result and isinstance(result, list) and all(line is not None for line in result):
        # Extract words
        for line in result:
            if line:  # Ensure that each line is valid and not None
                words.extend([word_info[1][0] for word_info in line])

    # Combine extracted words into a single paragraph
    paragraph = " ".join(words)

    return paragraph

# Function to extract text from the image without rotation or zooming
def extract(entity_name, image_link):
    image = load_image_from_url(image_link)

    # Extract text from the image
    paragraph = preprocess_and_extract_text(image)

    print(f"Extracted Paragraph: {paragraph}\n")

    return paragraph
```

```python
qa_pipeline = pipeline("question-answering", model="twmkn9/albert-base-v2-squad2")

```

```python
def answer(entity_name, paragraph, qa_pipeline):
    # Since we now have only one paragraph, we don't need multiple paragraphs
    extracted_text = 'input text: ' + paragraph

    # Formulate the query based on entity_name
    query = "what is the " + entity_name

    # Use the QA pipeline to find the answer
    result = qa_pipeline(question=query, context=extracted_text)

    return result['answer']
```

```python
import re

def extract_info(text):
    # Preprocess text to remove all non-numeric characters before the numerical part and strip any extra whitespace
    cleaned_text = re.sub(r'^[^\d]*', '', text).strip()
    
    # Extract the numerical prefix
    # match_prefix = re.match(r'(\d+\.?\d*)', cleaned_text)
    # prefix = match_prefix.group(1) if match_prefix else ''
    
    # # Extract the letter suffix
    # match_suffix = re.search(r'([a-zA-Z]+)$', cleaned_text)
    # suffix = match_suffix.group(1) if match_suffix else ''
    
    # Function to extract unit that occurs after the number or space
    def extract_unit_based_on_condition(text):
        # Regex to match units that appear right after a number or space following a number
        match_unit = re.search(r'(?<=\d)\s*([a-zA-Z]+)', text)
        return match_unit.group(1) if match_unit else ''
    
    # Extract unit using the new function
    extracted_unit = extract_unit_based_on_condition(cleaned_text)
    
    # Define regex patterns for various units
    units_patterns = {
        # Centimeters and variations
        r'(?<!\w)(\d+\.?\d*)\s*(cm|cms|centimetre|centimeter|centimeters|centimetres|c|centy|centis|c-m\b)': 'centimeter',
        
        # Inches and variations, including "
        r'(?<!\w)(\d+\.?\d*)\s*(in\b|inch\b|inches\b|inchs\b|"|inch\s|ins\b)': 'inch',
        
        # Millimeters and variations
        r'(?<!\w)(\d+\.?\d*)\s*(mm|mms|millimetre|millimeter|millimeters|millimetres|millis|milli)': 'millimeter',
        
        # Meters and variations (exclude confusion with mm)
        r'(?<!\w)(\d+\.?\d*)\s*(m\b(?!m)|metre|meter|meters|metres|mtrs\b|mts\b|meters\b|mtre|metr\b)': 'meter',
        
        # Feet and variations, including '
        r'(?<!\w)(\d+\.?\d*)\s*(ft\b|feet\b|foot\b|\'|feets|foots|foots\b)': 'foot',
        
        # Yards and variations
        r'(?<!\w)(\d+\.?\d*)\s*(yard\b|yd\b|yards\b|yds\b)': 'yard',

        # Kilometers and variations
        r'(?<!\w)(\d+\.?\d*)\s*(km|kms|kilometre|kilometer|kilometers|kilometres|klicks|klick)': 'kilometer',
        
        # Miles and variations
        r'(?<!\w)(\d+\.?\d*)\s*(mi\b|mile\b|miles\b|m\b(?!m)|miles\s|mil\b|mle\b)': 'mile',
        
        # Grams and variations
        r'(?<!\w)(\d+\.?\d*)\s*(g\b|grams\b|gram\b|gms\b|gm\b)': 'gram',
        
        # Kilograms and variations
        r'(?<!\w)(\d+\.?\d*)\s*(kg\b|kgs\b|kilogram\b|kilograms\b|kilo\b|kilos\b|kgm\b)': 'kilogram',
        
        # Liters and variations
        r'(?<!\w)(\d+\.?\d*)\s*(l\b|ltr\b|litre\b|liter\b|liters\b|litres\b|ltrs\b)': 'liter',
        
        # Milliliters and variations
        r'(?<!\w)(\d+\.?\d*)\s*(ml\b|millilitre\b|milliliters\b|millilitres\b|milliltrs\b)': 'milliliter',
        
        # Pounds (weight) and variations
        r'(?<!\w)(\d+\.?\d*)\s*(lb\b|lbs\b|pound\b|pounds\b|lbm\b|lbwt\b)': 'pound',
        
        # Ounces and variations
        r'(?<!\w)(\d+\.?\d*)\s*(oz\b|ounce\b|ounces\b)': 'ounce',
        
        # Gallons and variations
        r'(?<!\w)(\d+\.?\d*)\s*(gal\b|gallon\b|gallons\b|gals\b)': 'gallon',
        
        # Pints and variations
        r'(?<!\w)(\d+\.?\d*)\s*(pt\b|pint\b|pints\b)': 'pint',

        # Casual or informal representations of units (like "klicks" for kilometers, "kilos" for kilograms, etc.)
        r'(?<!\w)(\d+\.?\d*)\s*(klick\b|klicks\b|kilometers\b)': 'kilometer',
        r'(?<!\w)(\d+\.?\d*)\s*(kilo\b|kilos\b)': 'kilogram',
        r'(?<!\w)(\d+\.?\d*)\s*(w\b|watt\b|watts\b)': 'wattage',
        r'(?<!\w)(\d+\.?\d*)\s*(kw\b|kilowatt\b|kilowatts\b)': 'kilowatt',
        r'(?<!\w)(\d+\.?\d*)\s*(hp\b|horsepower\b)': 'horsepower',
        r'(?<!\w)(\d+\.?\d*)\s*(mw\b|megawatt\b|megawatts\b)': 'megawatt',
        r'(?<!\w)(\d+\.?\d*)\s*(v\b|volt\b|volts\b)': 'voltage',
        r'(?<!\w)(\d+\.?\d*)\s*(kv\b|kilovolt\b|kilovolts\b)': 'kilovolt',
        r'(?<!\w)(\d+\.?\d*)\s*(mv\b|millivolt\b|millivolts\b)': 'millivolt',
        r'(?<!\w)(\d+\.?\d*)\s*(µv\b|microvolt\b|microvolts\b)': 'microvolt'
    }

    # Check for unit type using the original patterns
    unit = ''
    for pattern, unit_type in units_patterns.items():
        if re.search(pattern, text, re.IGNORECASE):
            unit = unit_type
            break
    
    return {
        'prefix': prefix,
        'suffix': suffix,
        'unit': unit if unit else extracted_unit  # Prioritize pattern match, fallback to extracted unit
    }
```
